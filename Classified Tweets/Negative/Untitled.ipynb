```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

```python
def loadtext(filename):
    data_set = pd.read_csv(filename, names=["author id", "created_at", "geo", "id", "lang", "like_count", "quote_count", "reply_count", "retweet_count", "source", "tweet", "Processed Tweets", "Sentiment"],  encoding="ISO-8859-1" )
    data_set = data_set[["author id", "created_at", "geo", "id", "lang", "like_count", "quote_count", "reply_count", "retweet_count", "source", "tweet", "Processed Tweets", "Sentiment"]]    
    Ptweets= list(data_set["Processed Tweets"])
    Ptweets.pop(0)
    return Ptweets

def wordcloud(List):
    wc = WordCloud(max_words = 1000 , width = 1800 , height = 1000,collocations=False).generate(" ".join(List))
    plt.figure(figsize = (30,30))
    plt.imshow(wc)
def freq(l, filename):
    str2=[]
    for i in l:
        n=i.split(' ')
        for m in n:
            str2.append(m)
    counts = Counter(str2)
    with open(filename, "w") as output:
        output.write(str(counts))
    
```

```python
filename=["VaccineSentiment.csv", "PfizerSentiment.csv", "modernaSentiment.csv", "sinopharmSentiment.csv", "JohnsonSentiment.csv", "BoosterSentiment.csv", "antivaxSentiment.csv"]
finalList=[]
for i in range(7):
    rlist=loadtext("Negative_"+filename[i])
    print(len(rlist))
    finalList+=rlist
```

```python
freq(finalList, "comneg.txt")
```

```python
def freq2(nlist,words):
    str2=[]
    for i in nlist:
        n=i.split(' ')
        for m in n:
            if m in words:
                str2.append(m)
#             print(str2)
    return str2
words=["government","infections","pregnant","priority","eligible","republicans","republican","election","safety","safe","effective","effectiveness","strains","strain", "variant","variants","misinformation","mandatory","freedom","passport","healthcare"]
strlist=freq2(finalList, words)
```

```python
for i in range (len(strlist)):
    if strlist[i]=="variant":
        strlist[i]="variants"
    if strlist[i]=="effectiveness":
        strlist[i]="effective"
    if strlist[i]=="republican":
        strlist[i]="republicans"
    if strlist[i]=="safety":
        strlist[i]="safe"
    if strlist[i]=="strains":
        strlist[i]="strain"
        
```

```python
counts = Counter(strlist)
print(counts)
```

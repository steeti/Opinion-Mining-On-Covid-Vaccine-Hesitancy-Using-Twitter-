```python
import pandas as pd
def loadtext(filename):
    data_set = pd.read_csv(filename, names=["author id", "created_at", "geo", "id", "lang", "like_count", "quote_count", "reply_count", "retweet_count", "source", "tweet", "Processed Tweets", "Sentiment"],  encoding="ISO-8859-1" )
    data_set = data_set[["author id", "created_at", "geo", "id", "lang", "like_count", "quote_count", "reply_count", "retweet_count", "source", "tweet", "Processed Tweets", "Sentiment"]]
    f11= list(data_set["created_at"])
    f11.pop(0)
    return f11
```

```python
dataoutput=["VaccineSentiment.csv", "PfizerSentiment.csv", "modernaSentiment.csv", "sinopharmSentiment.csv", "JohnsonSentiment.csv", "BoosterSentiment.csv", "antivaxSentiment.csv"]
Naming=["Vaccine", "Pfizer", "moderna", "sinopharm", "Johnson", "Booster", "antivax"]
n=0
for i in dataoutput:
#     print(Naming[n]+"-Total")
    total=Counter(i)
    Neg="Negative_"+i
#     print(Naming[n]+"-Negative")
    negative=Counter(Neg)
    Pos="Positive_"+i
#     print(Naming[n]+"-Positive")
    positive=Counter(Pos)
    
    print(Naming[n]+"-Positive Percentage")
    for z in range (len(total)):
        if total[z]!=0:
            v=positive[z]/total[z]*100
            print(v)
        else:
            print(0)
    print(Naming[n]+"-Negative Percentage")
    for z in range (len(total)):
        if total[z]!=0:
            v=negative[z]/total[z]*100
            print(v)
        else:
            print(0)
    n+=1
        
    
    
```
